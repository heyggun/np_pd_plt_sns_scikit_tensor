## 1. 넘파이 생성

```python
A = np.array([1,2,3])
B = np.array([4,5,6])

print(A.shape, B.shape)
print(A.ndim, B.ndim)
```

## 3. 넘파이 형 변환(reshape)
- 머신러닝 코드 구현시, 넘파이 벡터(vector)를 행렬(matrix)로 변경하거나, 행렬을 다른 형상(shape) 행렬로 변경해야 할 때가 있음
- 넘파이의 reshape() 메서드 사용

```python
C = np.array([1,2,3])
print(C.shape)

C = C.reshape(1,3)
print(C.shape)
```

```python
D = np.array([[1,2,3],[4,5,6]])
print(D.shape)

D = D.reshape(3,2)
print(D.shape)
```

```python
A = np.array([[1,2,3,4],[10,20,30,40], [100,200,300,400]])
print(A.shape)
print(A)

B = A.reshape(-1,3)
print(B.shape)
print(B)
```

```python
C = A.reshape(-1,6)
print(C.shape)
print(C)
```

## 4. 넘파이 브로드캐스트(broadcast)

- 행렬의 사칙연산은 기본적으로 2개의 행렬 크기가 같은 경우에만 가능
- 넘파이(numpy)에서는 크기가 다른 두 행렬 간에도 사칙 연산을 할 수 있게 확장해주는 기능 : 브로드캐스트(broadcast)
- 차원이 작은 쪽이 큰 쪽에 맞추어 행을 단위로, 반복적으로 크기를 확장

```python
A = np.array([[1,2],[3,4]])
B = 5

print(A+B)
```

```python
C = np.array([[1,2],[3,4]])
D = np.array([4,5])

print(C+D)
```

- 차원이 낮은 스칼라(scalar)를 차원이 한 단계 높은 행렬과 동일한 크기로 확장해주는 기능 `넘파이 브로드캐스트(broadcast)`
- 오로지 `사칙 연산`에만 제공되며, `행렬곱 연산`에는 적용되지 않음

## 5. 넘파이 전치 행렬(transpose)
- 행렬의 행(row)과 열(column)이 바뀌는 전치(transpose)


```python
A = np.array([[1,2],[3,4],[5,6]])
B = A.T # 행렬 A에 대한 전치 행렬

print(A.shape)
print(B.shape)
print(A)
print(B)
```

`주의점! 벡터(vector)를 전치 행렬로 만들고자 할때는 기존 원본 벡터(vector)는 계속 벡터(vector)로 남아있음`

```python
C = np.array([1,2,3,4,5]) # 벡터에 대해 전치 행렬 수행
D = C.T

print(C.shape)
print(D.shape)
```

```python
E = C.reshape(1,5)
F = E.T

print(E.shape)
print(F.shape)
```

-> 벡터는 행과 열을 구분하지 않는 1차원이기 때문에 전치 행렬을 하더라도 열과 행을 구분할 수 없음

## 7. 행렬 원소 접근 (indexing / slicing)

## 8. 넘파이 이터레이터(iterator)
- 행렬의 모든 원소를 처음부터 끝까지 하나씩 가져오기 위해서 `이터레이터(iterator) 방법`사용

```python
A = np.array([[10,20,30,40], [50,60,70,80]])
print(A)
it = np.nditer(A, flags=['multi_index'], op_flags=['readwrite'])

# flags=['multi_index'] -> 이터레이터 생성 후 반복할 때 행렬(행,열) 형태 multi_index 형태로 동작해라
# op_flags['readwirte'] -> 이터레이터를 Read/Write 형태로 생성함


while not it.finished:
    idx = it.multi_index
    print('index->', idx, 'value->', A[idx])
    it.iternext()
```

```python
W = np.random.rand(1,1)
b = np.random.rand(1)

print(W.shape)
print(b.shape)
```

```python
def loss_func(xdata, tdata):
    y = np.dot(xdata, W) + b
    loss = np.sum((tdata-y)**2) / len(xdata)
    
    return loss
```
