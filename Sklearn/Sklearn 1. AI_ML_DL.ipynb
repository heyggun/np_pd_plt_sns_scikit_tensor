## (1) 인공지능(AI), 머신러닝(ML), 딥러닝(DL)




- `인공지능` : 인간이 생각하고 추론하는 능력을 컴퓨터로 구현한 것
- `머신러닝` : 인공지능을 구현하는 하나의 방법, 데이터의 특성과 패턴을 학습한 후에 학습 결과를 바탕으로 임의의 데이터에 대한 미래값을 예측 
즉, 하나의 알고리즘을 머신러닝으로 정의

## (2) 오차(error), 가중치(weight), 바이어스(bias)






- y = W*x + b 에서 정답값 t1, t2,t3,t4,t5,t6 에 대한 계산값 y1, y2, y3, y4, y5, y6 등 차이를 볼 수 있음
- 정답 t와 계산값 y의 차이 `오차(error 혹은 loss)`
- 여기서 오차 값이 크다는 것은 y = W*x +b 의 가중치 W와 바이어스 b 값이 잘못된 것
- 오차 값이 적다는 것은 y = W*x +b 함수의 가중치 W와 바이어스 b 값이 최적화 되었기 때문에 t-y 값인 오차(error)가 최소값이 된 것

- ML 시스템은 트레이닝 데이터의 오차(loss)값인 t-y 총합이 최소가 되서 미지의 데이터의 미래값을 잘 예측할 수 있도록
- `y= W*b 함수의 가중치 W와 바이어스 b 값을 찾는 과정이 필요함`

## (3) 손실함수(Loss function, 혹은 Cost function)



- 이러한 각각의 오차인 t-y를 모두 더해서 손실함수(loss function/Cost function)을 구할 수 있음 
- 이 때 t-y의 오차가 플러스(+)와 마이너스(-) 등이 동시에 존재해 합이 0이 되어, 그 값이 최소 오차값인지 판단학 ㅣ어려움
- 그래서, 손실함수에서는 `(t-y)^2` 사용
