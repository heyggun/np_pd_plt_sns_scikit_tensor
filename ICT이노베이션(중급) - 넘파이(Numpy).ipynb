## 8.  (iterator)
-         `(iterator) `

```python
A = np.array([[10,20,30,40], [50,60,70,80]])
print(A)
it = np.nditer(A, flags=['multi_index'], op_flags=['readwrite'])

# flags=['multi_index'] ->      (,)  multi_index  
# op_flags['readwirte'] ->  Read/Write  


while not it.finished:
    idx = it.multi_index
    print('index->', idx, 'value->', A[idx])
    it.iternext()
```

```python
W = np.random.rand(1,1)
b = np.random.rand(1)

print(W.shape)
print(b.shape)
```

```python
def loss_func(xdata, tdata):
    y = np.dot(xdata, W) + b
    loss = np.sum((tdata-y)**2) / len(xdata)
    
    return loss
```

```python
xdata = np.array([1,2,3,4,5]).reshape(5,1)
tdata = np.array([3,4,5,6,7]).reshape(5,1)

ret = loss_func(xdata, tdata)
print(ret)
```

```python
x = np.arange(0,15,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,5))
plt.title('sin(x)/cos(x)')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid()
plt.xlim(0,15)
plt.ylim(-1,1)

plt.plot(x,y1, label='sin(x)', color='r', ls='--')
plt.plot(x,y2, label='cos(x)', color='g', ls='--')

plt.legend(loc='best')
plt.show()
```
