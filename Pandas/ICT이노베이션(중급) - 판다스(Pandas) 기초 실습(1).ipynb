### 9. 데이터프레임 `피벗테이블(Pivot Table)`
- 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터 조회하여 펼쳐 놓은 것
- 왼쪽에 나타나는 인덱스 행 인덱스, 상단에 나타나는 인덱스 열 인덱스

```python
pd.pivot_table(data, index='소속사', columns='혈액형', values='키')
```

- index : 행 인덱스
- columns : 열 인덱스
- values : 조회하고 싶은 값

```python
pd.pivot_table(data, index='소속사', columns='혈액형', values='브랜드평판지수', aggfunc=np.mean)
```

### 10. 데이터프레임 `GroupBy`
- 그룹으로 묶어 분석할 때
- 예) 소속사별 키의 평균, 성별 키의 평균등과 같이 특정 그룹별 통계 데이터 성질 확인

```python
data.groupby('소속사')
```

- count() 개수
- sum() 합계
- mean() 평균
- var() 분산
- std() 표준편차
- min() / max() 최소값/최대값

```python
data.groupby('소속사').count()
```

- 산술통계는 자동으로 산술통계가 가능한 열만 출력됨

```python
data.groupby('그룹').mean()
```

```python
data.groupby('성별').sum()
```

```python
# 특정 열만 출력하고 싶을 때
data.groupby('혈액형')['키'].mean()
```

### 11. `Multi-Index` : 복합 인덱스
- 행 인덱스를 복합적으로 구성하고 싶은 경우, 인덱스를 리스트로 만들어 줌

```python
data.groupby(['혈액형', '성별']).mean()
```

#### 11-1. Multi-index로 된 데이터프레임을 피벗테이블 형태로 다시 변환

```python
df2 = data.groupby(['혈액형','성별']).mean()
df2
```

```python
df2.unstack('혈액형')
```

```python
df2.unstack('성별')
```

#### 11-2. 인덱스 초기화

```python
df2 = df2.reset_index()
df2
```
