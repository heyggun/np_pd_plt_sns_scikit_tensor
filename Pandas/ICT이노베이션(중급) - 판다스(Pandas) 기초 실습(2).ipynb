## 3. 데이터프레임 `병합(merge)`

- concat과 merge는 단순 합치는 목적과 특정 기준(index)로 합치느냐에 따라 용도가 다름
- concat : row나 column 기준으로 단순하게 이어 붙이기
- merge : 특정 고유한 키 (unique id)값을 기준으로 병합 

```python
df1.head(3)
```

```python
df2.head(3)
```

- df1, df2는 '이름' column이 겹침 
- '이름' column으로 병합 가능

''' pd.merge(left, right, on='기준 column', how='left') '''

- `left`, `right`는 병합할 두 DataFrame
- `on` 에는 병합의 기준이 되는 column
- `how`에는 'left', 'right', 'inner', 'outer' 라는 4가지 병합 방식 중 한가지 선택

#### 3-1. left, right 방식

```python
# df2에서 5개의 데이터 고의 drop

df_right = df2.drop([1,3,5,6,8], axis=0).reset_index(drop=True)
df_right
```

```python
pd.concat([df1, df_right], axis=1)
```

```python
pd.merge(df1, df_right, on='이름', how='left')
```

```python
pd.merge(df1, df_right, on='이름', how='right')
```

- `inner` : 두 DataFrame에 모두 키 값이 존재하는 경우에만 병합
- `outer` : 하나의 DataFrame에 키 값이 존재하는 경우 모두 병합(`outer` 방식에서 없는 값은 NaN으로 대입)

#### 4-1. apply

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```python
df.loc[df['성별']=='남자', '성별'] = 1
df.loc[df['성별']=='여자', '성별'] = 0
df
```

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
df['성별'].value_counts()
```

```python

def change_sex(x):
    
    if x == '남자':
        sex = 1
        
    elif x == '여자':
        sex = 0
        
    else :
        sex = -1
        
    return sex

df['성별_new'] = df['성별'].apply(change_sex)
df
```
